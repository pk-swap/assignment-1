### Data visualization

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset(r"C:\Users\palla\OneDrive\Desktop\hwfod\Health_heart_experimental.csv")
df.head(10)
```

### line plot

```python

plt.hist(df['weightKg'])
```

###### In the above graph, BMI stays almost the same at all ages and does not change much as age increases.

### barplot

```python

sns.barplot(x='sex',y='BMI',data=df)
```

###### In the above bar chart, BMI is almost the same for both males and females which shows sex does not affect BMI much.

### piechart

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

###### The pie chart shows that both groups, female (0) and male (1), have nearly equal numbers of patients, which means the dataset is balanced between the two.

### scatterpplot

```python

sns.scatterplot(x='weightKg', y='BMI', data=df, hue='sex', )
plt.title('weightKg vs HR')
plt.xlabel('weightKg')
plt.ylabel('BMI')
plt.show()
```

###### The scatter plot shows the relationship between weight and BMI. Patients with lower weight have BMI values in a narrow range, while higher weights show a wider spread of BMI. 

```python
sns.boxplot(df['BMI'])
plt.title("Box Plot: BMI Distribution")
plt.show()
```

###### The box plot shows that most patients have a BMI between 24 and 35, with the average around 29. This means many patients fall in the region are overweight . 

```python
plt.hist(df['weightKg'])
```

###### The histogram is right-skewed, which mean most patients weigh less than 80kg. The common weight range is 5060kg, and only a few patients weigh over 100kg, which is good since very high weight is injurious health.
